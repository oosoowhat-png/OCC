```python
from pyspark import SparkContext

sc = SparkContext.getOrCreate()

# 1. Load data
rdd = sc.textFile("EPL.csv")
header = rdd.first()
rdd = rdd.filter(lambda row: row != header).map(lambda row: row.split(","))

# Columns: 0=home_team, 1=away_team, 2=home_goals, 3=away_goals, 4=result, 5=season

# 2. Season with highest/lowest goals
season_goals = rdd.map(lambda x: (x[5], int(float(x[2])) + int(float(x[3])))) \
                  .reduceByKey(lambda a, b: a + b)
highest_season = season_goals.max(key=lambda x: x[1])
lowest_season = season_goals.min(key=lambda x: x[1])

print("\n=== Season Goal Stats ===")
print(f"Highest goals season: {highest_season[0]} with {highest_season[1]} goals")
print(f"Lowest goals season: {lowest_season[0]} with {lowest_season[1]} goals")

# 3. Team with highest average goals per season
team_goals = rdd.flatMap(lambda x: [
    ((x[0], x[5]), int(float(x[2]))),  # (team, season) -> home goals
    ((x[1], x[5]), int(float(x[3])))   # (team, season) -> away goals
])

# Total goals per team-season
team_season_total = team_goals.reduceByKey(lambda a, b: a + b)

# (team -> (total_goals, set_of_seasons))
team_total = team_season_total.map(lambda x: (x[0][0], (x[1], {x[0][1]}))) \
                              .reduceByKey(lambda a, b: (a[0] + b[0], a[1] | b[1]))

team_avg = team_total.map(lambda x: (x[0], x[1][0] / len(x[1][1])))
best_team = team_avg.max(key=lambda x: x[1])

print("\n=== Best Team by Avg Goals ===")
print(f"{best_team[0]} with average {best_team[1]:.2f} goals per season")

# 4. Manchester United probabilities
mu_matches = rdd.filter(lambda x: x[0] == "Manchester United" or x[1] == "Manchester United")
total_matches = mu_matches.count()

mu_wins = mu_matches.filter(
    lambda x: (x[0] == "Manchester United" and x[4] == "H") or
              (x[1] == "Manchester United" and x[4] == "A")
).count()

mu_losses = mu_matches.filter(
    lambda x: (x[0] == "Manchester United" and x[4] == "A") or
              (x[1] == "Manchester United" and x[4] == "H")
).count()

mu_draws = mu_matches.filter(lambda x: x[4] == "D").count()

print("\n=== Manchester United Probabilities ===")
print(f"P(Win)  = {mu_wins / total_matches:.3f}")
print(f"P(Loss) = {mu_losses / total_matches:.3f}")
print(f"P(Draw) = {mu_draws / total_matches:.3f}")
```

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, avg, count, sum as _sum, desc
import matplotlib.pyplot as plt
import pandas as pd

# Initialize
sc = SparkContext.getOrCreate()
spark = SparkSession(sc)

# Load CSV
df = spark.read.csv("EPL.csv", header=True, inferSchema=True)

print("=== Dataset Schema ===")
df.printSchema()

print("=== First 5 Rows ===")
df.show(5)

# ---------------------------------------------------------
# 1. Basic Stats
# ---------------------------------------------------------
print("\n=== Basic Summary ===")
df.describe().show()

print("\nNumber of Seasons:", df.select("season").distinct().count())
print("Number of Teams:", df.select("home_team").distinct().count())

# ---------------------------------------------------------
# 2. Goals per Season
# ---------------------------------------------------------
season_goals = df.withColumn("total_goals", col("home_goals")+col("away_goals")) \
                 .groupBy("season").agg(_sum("total_goals").alias("goals")) \
                 .orderBy("season")

season_goals_pd = season_goals.toPandas()

plt.figure(figsize=(10,5))
plt.bar(season_goals_pd["season"], season_goals_pd["goals"], color="skyblue")
plt.title("Total Goals per Season")
plt.xlabel("Season")
plt.ylabel("Goals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

highest_season = season_goals.orderBy(desc("goals")).first()
lowest_season = season_goals.orderBy("goals").first()
print(f"Highest goals season: {highest_season['season']} ({highest_season['goals']})")
print(f"Lowest goals season: {lowest_season['season']} ({lowest_season['goals']})")

# ---------------------------------------------------------
# 3. Team with Highest Avg Goals per Season
# ---------------------------------------------------------
home = df.groupBy("season","home_team").agg(_sum("home_goals").alias("goals")) \
         .withColumnRenamed("home_team","team")

away = df.groupBy("season","away_team").agg(_sum("away_goals").alias("goals")) \
         .withColumnRenamed("away_team","team")

team_season_goals = home.union(away)

team_stats = team_season_goals.groupBy("team") \
    .agg(_sum("goals").alias("total_goals"),
         count("season").alias("seasons_played")) \
    .withColumn("avg_goals", col("total_goals")/col("seasons_played")) \
    .orderBy(desc("avg_goals"))

team_stats_pd = team_stats.toPandas().head(10)

plt.figure(figsize=(10,5))
plt.barh(team_stats_pd["team"], team_stats_pd["avg_goals"], color="orange")
plt.title("Top 10 Teams by Avg Goals per Season")
plt.xlabel("Avg Goals per Season")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

best_team = team_stats.first()
print(f"üèÜ Best Team: {best_team['team']} with {best_team['avg_goals']:.2f} avg goals/season")

# ---------------------------------------------------------
# 4. Manchester United Probability Analysis
# ---------------------------------------------------------
mu = df.filter((col("home_team")=="Manchester United") | (col("away_team")=="Manchester United"))
total = mu.count()

wins = mu.filter(
    ((col("home_team")=="Manchester United") & (col("result")=="H")) |
    ((col("away_team")=="Manchester United") & (col("result")=="A"))
).count()

losses = mu.filter(
    ((col("home_team")=="Manchester United") & (col("result")=="A")) |
    ((col("away_team")=="Manchester United") & (col("result")=="H"))
).count()

draws = mu.filter(col("result")=="D").count()

print("\n=== Manchester United Outcome Probabilities ===")
p_win, p_loss, p_draw = wins/total, losses/total, draws/total
print(f"P(Win)  = {p_win:.3f}")
print(f"P(Loss) = {p_loss:.3f}")
print(f"P(Draw) = {p_draw:.3f}")

# Pie Chart for MU Results
plt.figure(figsize=(6,6))
plt.pie([wins, losses, draws], labels=["Win","Loss","Draw"], autopct="%1.1f%%", colors=["green","red","gray"])
plt.title("Manchester United Results Distribution")
plt.show()

# ---------------------------------------------------------
# 5. Extra Analytics
# ---------------------------------------------------------

# (a) Result distribution overall
result_dist = df.groupBy("result").count().toPandas()
plt.figure(figsize=(6,6))
plt.pie(result_dist["count"], labels=result_dist["result"], autopct="%1.1f%%", colors=["blue","orange","gray"])
plt.title("Overall Match Result Distribution (H/A/D)")
plt.show()

# (b) Top 5 Highest Scoring Matches
goal_fests = df.withColumn("total_goals", col("home_goals")+col("away_goals")) \
               .orderBy(desc("total_goals")).limit(5)
goal_fests.show()
```
